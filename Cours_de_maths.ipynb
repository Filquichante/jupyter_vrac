### On cherche à représenter la fréquence d'apparition du 1 au fur et à mesure que l'on lance plus de dés.

```python
import random
from matplotlib import pyplot as plt

n=1000
c = 0
liste = []
for x in range(1,n+1):
  i = random.randint(1,4)
  if i==1:
    c+=1
  liste.append(c/x)
print(c/n)
```

```python
plt.plot(range(1,n+1), liste, '-')
plt.grid()
plt.title("Fréquence d'apparition du 1")
plt.show()
```

###On essaie de représenter la fluctuation de la fréquence d'apparition du 1 en fonction de la taille de l'échantillon.

```python
tailles = [100,1000,10000]
couleurs = ["blue","green","yellow"]
for x in range(0,3):
  N = 50
  M = tailles[x]
  liste_listes = []
  for i in range(1,N+1):
    c = 0
    liste_frequence = []
    for j in range(1,M+1):
      a = random.randint(1,4)
      if a == 1:
        c += 1
      liste_frequence.append(c/j)
    b = c/M
    liste_listes.append(b)
  plt.scatter(range(N),liste_listes, s=10, color=couleurs[x])
plt.axis(ymin = 0, ymax = 1)
plt.grid()
plt.show()
```
