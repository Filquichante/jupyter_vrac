```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.arange(-20,20,0.1)
def f(x):
  return 2*x**2 - 12*x + 3

y = f(x)
plt.plot(x,y)

def derf(x,f):
  h = 0.00001
  return (f(x+h)-f(x))/h

def tangente(i,f):
  x = np.arange(-20,20,0.1)
  for a in i:
    plt.plot(x, derf(a,f)*(x-a)+f(a))

tangente(range(-20,20),f)
```
